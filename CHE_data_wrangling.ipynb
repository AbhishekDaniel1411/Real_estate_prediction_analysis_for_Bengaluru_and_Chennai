```python
import pandas as pd
import numpy as np
```

```python
che_df = pd.read_csv("Chennai houseing sale.csv")

che_df.head()
```

```python
che_df.shape
```

```python
che_df.columns
```

```python
che_df.drop(['PRT_ID','DATE_SALE', 'DIST_MAINROAD','N_ROOM', 'SALE_COND', 
             'PARK_FACIL', 'DATE_BUILD','BUILDTYPE', 'UTILITY_AVAIL', 'STREET', 
             'MZZONE', 'QS_ROOMS','QS_BATHROOM', 'QS_BEDROOM', 'QS_OVERALL', 
             'REG_FEE', 'COMMIS'], axis = 1, inplace = True)
```

```python
che_df.head()
```

```python
che_df.isnull().sum()
```

```python
che_df['N_BATHROOM'].replace(np.nan, che_df['N_BATHROOM'].median(), inplace= True)
```

```python
che_df.dropna(inplace = True)
```

```python
che_df.isnull().sum()
```

```python
che_df.info()
```

```python
che_df['SALES_PRICE'] = np.round(che_df['SALES_PRICE']/100000, decimals = 2)
```

```python
che_df
```

```python
che_df['INT_SQFT'] = che_df['INT_SQFT'].astype(float)
```
