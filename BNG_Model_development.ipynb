```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
bng = pd.read_csv('BNG_feature_eng.csv')
```

```python
bng.head()
```

```python
One_hot_encoder = pd.get_dummies(bng.location)

One_hot_encoder
```

```python
One_hot_encoder = One_hot_encoder.astype(int)
```

```python
bng = pd.concat([bng.drop('location', axis = 1), One_hot_encoder.drop('Others', axis = 1)], axis = 1)
```

```python
bng.head()
```

```python
#df of Independent features
X = bng.drop(['price', 'price_per_sqft'], axis = 1)

X.head()
```

```python
Y = bng['price']

Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.2, random_state=4)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(Xtrain, Ytrain)
```

```python
lr.score(Xtest, Ytest)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
```

```python
cv = ShuffleSplit(n_splits=4, test_size=0.2, random_state=1)
```

```python
cross_val_score(LinearRegression(), X, Y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
```

```python
def gridsearchcv_models(X, Y):
    algos = {
        'linear_regression':{
            'model': LinearRegression(),
            'params': {
                #'normalize': [True, False]
            }
        },
        'Ridge':{
            'model': Ridge(),
            'params': {
                'alpha': [0.1, 1, 10, 100],
                'solver': ['svd', 'lbfgs']
            }
        },
        'lasso':{
            'model': Lasso(),
            'params': {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree':{
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['mse', 'friedman_mse'],
                'splitter': ['best', 'random'] 
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=1)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv = cv)
        gs.fit(X,Y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    return pd.DataFrame(scores, columns=['model', 'Best Score', 'Best Parameters'])

gridsearchcv_models(X, Y)
```

```python
params_lasso = {
    'alpha':[1,2],
    'selection': ['random', 'cyclic']
}

gs1 = GridSearchCV(Lasso(), params_lasso, cv = cv)
gs1.fit(X,Y)
```

```python
gs1.best_score_
```

```python
gs1.best_params_
```

```python
params_ridge = {
    'alpha': [0.1, 1, 10, 100],
    'solver': ['svd', 'auto']
}

gs2 = GridSearchCV(Ridge(), params_ridge, cv = cv)
gs2.fit(X,Y)
```

```python
gs2.best_score_
```

```python
gs2.best_params_
```

```python
params_tree = {
    'criterion': ['poisson', 'friedman_mse'],
    'splitter': ['best', 'random']
}

gs3 = GridSearchCV(DecisionTreeRegressor(), params_tree, cv = cv)
gs3.fit(X,Y)
```

```python
gs3.best_params_
```

```python
gs3.best_score_
```

```python
params_lr = {
    'fit_intercept': [True, False],
    'positive': [True, False]
}

gs4 = GridSearchCV(LinearRegression(), params_lr, cv = cv)
gs4.fit(X,Y)
```

```python
gs4.best_params_
```

```python
gs4.best_score_
```

```python
df = pd.DataFrame(
    {'Model': ['Lasso', 'Ridge', 'Decision Tree Regressor', 'Linear Regression'],
    'Best Parameters': [gs1.best_params_, gs2.best_params_, gs3.best_params_, gs4.best_params_],
    'Best Score': [gs1.best_score_, gs2.best_score_, gs3.best_score_, gs4.best_score_]}
)

df
```

```python
def predict_price(location, sqft, bed, bath):
    loc_index = np.where(X.columns == location)[0][0]
    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bed
    x[2] = bath
    if loc_index >= 0:
        x[loc_index] = 1
    return gs4.predict([x])[0]
```
